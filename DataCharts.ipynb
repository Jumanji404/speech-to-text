```python
from matplotlib import pyplot as plt
```

```python
def plotAgeData():
    ageLabels=['<19','19-29','30-39','40-49','60-69','other']
    sizes=[5,27,51,8,1,8]
    percentSizes=[str(size)+'%' for size in sizes]
    plt.pie(sizes,labels=ageLabels)
    plt.legend(bbox_to_anchor=(1.5,0.75),labels=percentSizes)
    plt.title('Data By Age Distribution')
    plt.show()
```

```python
def plotSexData():
    ageLabels=['masculin','feminin','other']
    sizes=[72,20,8]
    percentSizes=[str(size)+'%' for size in sizes]
    plt.pie(sizes,labels=ageLabels)
    plt.legend(bbox_to_anchor=(1.5,0.75),labels=percentSizes)
    plt.title('Data By Sex Distribution')
    plt.show()
```

```python
plotAgeData()
```

```python
plotSexData()
```
